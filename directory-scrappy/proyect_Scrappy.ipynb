```python
import urllib
import re
import json
import os
import csv
import re
from datetime import date
from bs4 import BeautifulSoup


def download(url, user_agent='Mozilla/5.0', num_retries=2):
    #Descarga una url. En caso de existir errores errores [500 600], intentar nuevamente.
    #Función adaptada desde (Lawson Richard, 2015)
    print('Descargando: ', url)
    headers = {'User-agent': user_agent}
    request = urllib.request.Request(url, headers=headers)
    try:
        html = urllib.request.urlopen(request).read()
    except urllib.error.URLError as e:
        print('Error descargando:', e.reason)
        html = None
        if num_retries > 0:
            if hasattr(e, 'code') and 500 <= e.code < 600:
                return download(url, user_agent, num_retries-1)
    return html
```

```python
url = 'https://www.utpl.edu.ec/directorio/index.php?ban=1&id_dep=7&id_citte=7&bandera=0&nom_dep=COORDINACIÓN%20DE%20CÁTEDRAS%20UNESCO%20-%20UTPL'
# Page encoding

html = download(url) 

#page = requests.get(url)
#page.encoding = "utf-8"
#page

soup = BeautifulSoup(html, "html.parser")

# Print HTMl DOC
soup

# To parse the page's text
#soup = BeautifulSoup(page.text, "html.parser")

# Print HTMl DOC
soup

items = soup.findAll("div", {"class": "search_result_item line bgwhite mb1 borderb2 borderlightgrey"});
l1 = []
l2 = []

for i in items:
    l1.append(i.find("p").text)
    l2.append(i.find("strong").text)
print(len(l1))
print(len(l2))
```
